```python
%matplotlib inline
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
from sklearn.metrics import confusion_matrix
import pandas as pd
```

```python
from mnist import MNIST
data = MNIST(data_dir="data/MNIST/")
```

##### All of these data.x_train, data.x_test, data.x_val are numpy array.

```python
print(data.x_train.shape)
print(data.y_train.shape)
print(data.x_test.shape)
print(data.y_test.shape)
print(data.x_val.shape)
print(data.y_val.shape)
print(data.y_train_cls[0:5])
print(data.y_train[0:5])
print(data.y_test_cls[0:5])
print(data.y_test[0:5])
print(data.y_val_cls[0:5])
print(data.y_val[0:5])
```

```python
x_train = data.x_train
y_train = data.y_train
x_test = data.x_test
y_test = data.y_test
y_test_cls = data.y_test_cls
```

```python
### PART 2 - Create ANN and training it

import keras
from keras.models import Sequential
from keras.layers import Dense

# initialize ANN ( creating skeleton)
classifier = Sequential()

# adding input layer and first hidden layer to skeleton
classifier.add(Dense(output_dim = 196, init = 'uniform', activation = 'relu', input_dim = 784 ))

# adding one more hidden layer
classifier.add(Dense(output_dim = 10, kernel_initializer = 'uniform', activation = 'relu', input_dim = 196))

# adding output layer
classifier.add(Dense(units = 10, kernel_initializer = 'uniform', activation = 'softmax'))

# compile the ANN that is, add optimizer(SGD) and loss function
classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])

# now, the ANN is ready for training
# Training/fitting ANN to training set
classifier.fit(x_train, y_train, batch_size = 100, epochs = 100)


### PART 3 - Making predictions and Evaluating the model

# predicting testset results
y_pred = classifier.predict(x_test)
```

```python
y_pred_cls = np.argmax(y_pred, axis=1)
```

```python
y_pred_cls
```

```python
# Calculating accuracy
size = y_test.shape[0]
correctely_predicted = 0
for i in range(size):
    if (y_pred_cls[i] == y_test_cls[i]):
        correctely_predicted += 1
    
print("Accuracy:")
print(correctely_predicted / size)
```
